# Convex Hull using Quickhull
The basic idea is given a set of points, find the wrapping (Convex Hull is also reffered to as a gitf wrapping algorithm) points and reconstruct a 3D mesh of the shape

### Imports

```python
# Import matplotlib to draw in 3D
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Import numpy for numeric computations
import numpy as np

# Use scipy for Quickhull ConvexHull implementation
from scipy.spatial import ConvexHull
```

## Generate a random(ish) set of points

```python
sphere_points = np.random.randn(1500, 3)

fig, ax = plt.subplots(1, 1, subplot_kw={'projection':'3d', 'aspect':'equal'})
ax.scatter(sphere_points[:, 0], sphere_points[:, 1], sphere_points[:, 2], s=5, c='r', zorder=10)
```

## Calculate the Convex Hull and draw the points

```python
# Calculate the convex hull (i.e wrapping points) for the set of points
hull = ConvexHull(sphere_points)

# Plot it (red is the convex hull points, i.e the wrapping points and grey is the rest)
fig, ax = plt.subplots(1, 1, subplot_kw={'projection':'3d', 'aspect':'equal'})

# Plot all the points
ax.scatter(sphere_points[:, 0], sphere_points[:, 1], sphere_points[:, 2], s=2, c='grey', zorder=10)

# Plot the convex hull points in red
ax.scatter(sphere_points[hull.vertices, 0], 
           sphere_points[hull.vertices, 1], 
           sphere_points[hull.vertices, 2], 
           s=2, c='r', zorder=10)
```

## Use the results from the Convex Hull to draw a triangular surface

```python
fig, ax = plt.subplots(1, 1, subplot_kw={'projection':'3d', 'aspect':'equal'})
ax.plot_trisurf(sphere_points[:, 0], 
                sphere_points[:, 1], 
                sphere_points[:, 2], 
                triangles=hull.simplices, 
                cmap=plt.cm.Spectral)
```
